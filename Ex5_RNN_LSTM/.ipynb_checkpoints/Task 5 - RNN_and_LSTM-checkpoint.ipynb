# <span style="color: #1BB5B7;font-size: 25pt">2.  Long Short Term Memory networks with Keras - Netflix Stock Prices Prediction</span>

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,LSTM,Flatten,Dropout,BatchNormalization
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
df = pd.read_csv('NFLX.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df=df['Close']
df
```

```python
dates = df.index
dates = np.array(dates)

q_80 = int(len(df) * .8)
q_90 = int(len(df) * .9)

train_data = df[:q_80]
valid_data = df[q_80:q_90]
test_data = df[q_90:]
print("Train data shape:",train_data.shape)
print("Valid data shape:",valid_data.shape)
print("Test data shape:",test_data.shape)
#print(train_data)
```

```python
train_data = train_data.values.reshape(-1,1)
valid_data = valid_data.values.reshape(-1,1)
test_data = test_data.values.reshape(-1,1)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
train_data_scaled = scaler.fit_transform(train_data)
valid_data_scaled =scaler.transform(valid_data)
test_data_scaled =scaler.transform(test_data)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(dates[:q_80], train_data_scaled)
plt.plot(dates[q_80:q_90], valid_data_scaled)
plt.plot(dates[q_90:], test_data_scaled)
plt.xticks(rotation=45) 
plt.show()
```

```python
def create_seq(data,time_step=60):
    X=[]
    y=[]
    for i in range(len(data)-time_step-1):
        X.append(data[i:(i+time_step)])
        y.append(data[i+time_step])
    return np.array(X), np.array(y)
```

```python
time_step = 30
X_train, y_train = create_seq(train_data_scaled, time_step)
X_valid, y_valid = create_seq(valid_data_scaled, time_step)
X_test, y_test = create_seq(test_data_scaled, time_step)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.reshape(-1).shape)
print("X_valid shape:", X_valid.shape)
print("y_valid shape:", y_valid.reshape(-1).shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.reshape(-1).shape)
```

# <span style="color: #5EC64A;font-size: 25pt">1) simple model </span>

```python
time_step = 30

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(time_step, 1)),
    Flatten(),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(4, activation='relu'),
    Dense(1, activation='linear')
])

model.summary()
model.compile(loss='mse', optimizer=Adam(learning_rate=0.001), metrics=['mae'])

history = model.fit(X_train, y_train, validation_data=(X_valid, y_valid), 
                    epochs=100)

#best_model = tf.keras.models.load_model(checkpoint_filepath)
#best_model.compile(loss='mse',optimizer=Adam(learning_rate=0.001),metrics=['mae'])
#best_model.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=20)

test_loss, test_mae = model.evaluate(X_test, y_test, verbose=2)
print(f'Test Loss: {test_loss:.4f}, Test MAE: {test_mae:.4f}')
```

```python
def test_model(model, x_test, y_test, history, X_train, y_train):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['mae']) 
    plt.plot(history.history['val_mae']) 
    plt.title('Mean Absolute Error (MAE)') 
    plt.ylabel('MAE value') 
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper right')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss (MSE)') 
    plt.ylabel('MSE value')  
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper right')

    plt.tight_layout()
    plt.show()
    
    
    train_predictions = model.predict(X_train).flatten()
    train_dates = dates[time_step:q_80-1]

    plt.figure(figsize=(14, 7))
    plt.plot(train_dates, train_predictions[:len(train_dates)], label='Training Predictions')
    plt.plot(train_dates, y_train, label='Training Observations')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()
    

    loss, mae = model.evaluate(x_test, y_test)
    print('Mean Absolute Error on test set:', mae)
    
    loss, mae = model.evaluate(X_train, y_train)
    print('Mean Absolute Error on train set:', mae)
```

```python
test_model(model, X_test, y_test, history, X_train, y_train)
```
